```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
'''
    Loading csv files for structured data and report data
'''
# structured data
structured_file_path = "../output/structured_term_freq.csv"
structured_df = pd.read_csv(structured_file_path)
structured_terms = structured_df["Label"]
structured_frequencies = structured_df["Frequency"]

# report data
report_file_path = "../output/report_term_freq.csv"
report_df = pd.read_csv(report_file_path)
report_terms = report_df["Term"]
report_frequencies = report_df["Frequency"]
```

```python
'''
    Wordcloud for structured data
'''
word_freq = dict(zip(structured_terms, structured_frequencies))
wordcloud = WordCloud(
    width=800, 
    height=400, 
    background_color="white", 
    colormap="viridis"
).generate_from_frequencies(word_freq)

# Plot the word cloud
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Terms", fontsize=16)
plt.show()
```

```python
'''
    Wordcloud for report data
'''
word_freq = dict(zip(report_terms, report_frequencies))
wordcloud = WordCloud(
    width=800, 
    height=400, 
    background_color="white", 
    colormap="viridis"
).generate_from_frequencies(word_freq)

# Plot the word cloud
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Terms", fontsize=16)
plt.show()
```
