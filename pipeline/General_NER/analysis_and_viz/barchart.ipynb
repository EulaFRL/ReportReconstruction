```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
'''
    Structured Bar Chart
'''
structured_df = pd.read_csv("../output/structured_term_freq.csv")
top_n = 20
structred_top_terms = structured_df.sort_values(by="Frequency", ascending=False).head(top_n)

plt.figure(figsize=(8, 6))
plt.bar(structred_top_terms["Label"], structred_top_terms["Frequency"])
plt.xticks(rotation=45, ha="right")
plt.xlabel("Terms")
plt.ylabel("Frequency")
plt.title("Top 20 Terms by Frequency")
plt.show()
```

```python
'''
    Report Bar Chart
'''
report_df = pd.read_csv("../output/report_term_freq.csv")
top_n = 20
report_top_terms = report_df.sort_values(by="Frequency", ascending=False).head(top_n)

plt.figure(figsize=(8, 6))
plt.bar(report_top_terms["Term"], report_top_terms["Frequency"])
plt.xticks(rotation=45, ha="right")
plt.xlabel("Terms")
plt.ylabel("Frequency")
plt.title("Top 20 Terms by Frequency")
plt.show()
```
